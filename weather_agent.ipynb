```python
import os
from langchain.agents import create_agent
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
# 1. Define your tool
def get_weather(city: str) -> str:
    """Get weather for a given city."""
    return f"It's always sunny in {city}!"
```

```python
# 2. Set your Gemini API key here
from google.colab import userdata
api_key=userdata.get('gemini_key')
```

```python
# 3. Initialize the actual Gemini model object
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",api_key=api_key,
    temperature=0
)
```

```python
# 4. Create the agent using the Gemini model
agent = create_agent(
    model=llm,
    tools=[get_weather],
    system_prompt="You are a helpful assistant",
)
```

```python
# 5. Run the agent
result = agent.invoke(
    {"messages": [{"role": "user", "content": "What's the todays weather in Kolhapur,Maharashtra, India?"}]}
)
```

```python
# 6. Print the answer
print(result["messages"][-1].content)
```
